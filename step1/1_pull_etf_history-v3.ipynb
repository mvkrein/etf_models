```python
import pandas as pd
import numpy as np
import os
import sys
import fix_yahoo_finance as yf 
from datetime import datetime
pd.set_option('display.max_rows', 200)
pd.set_option('display.max_columns', 100)
pd.set_option('display.width', 1000)
```

```python
data_path = '/home/mvkrein/etf_model/data'
etf_file = os.path.join(data_path,'Fidelity_no_commission_ETFs.csv')
etf_list = pd.read_csv(etf_file)
```

```python
etf_list.sort_values(by='Symbol',inplace=True)
etf_list.reset_index(inplace=True, drop=True)
etf_list
```

```python
import fix_yahoo_finance as yf 

for i,etf in enumerate(etf_list['Symbol']):
    d = yf.download(etf,'2012-01-01','2018-09-20')
    d['Symbol'] = etf
    if i == 0:
        result = d
    else:
        result = pd.concat([result,d],axis=0)
```

```python
result.reset_index(drop=False,inplace=True)
result
```

```python
df_sym_min_date = pd.DataFrame(result.groupby('Symbol')['Date'].min())
```

```python
df_sym_min_date.reset_index(drop=False,inplace=True)
df_sym_min_date
```

```python
etf_6yr_list = df_sym_min_date.loc[df_sym_min_date['Date']=='2012-01-03',:]
etf_6yr_list = pd.DataFrame(etf_6yr_list['Symbol'])
etf_6yr_list.reset_index(drop=True,inplace=True)
etf_6yr_list
```

```python
etf_6yr_hist = result.merge(etf_6yr_list,how='inner',on='Symbol')
etf_6yr_hist
```

```python
etf_6yr_hist.groupby('Symbol')['Date'].count()
```

```python
etf_6yr_hist=etf_6yr_hist.loc[etf_6yr_hist['Symbol'] != 'AGT',]
```

```python
etf_6yr_hist.groupby('Symbol')['Date'].count()
```

```python
etf_final = etf_6yr_hist.loc[:,['Date','Symbol','Adj Close','Volume']]
```

```python
etf_final.columns=['Date','sym','p','v']
etf_final
```

```python
etf_data_file = os.path.join(data_path,'Fidelity_data_from_20120101_new.csv')
etf_final.to_csv(etf_data_file,index=False)
```

```python
data_path = '/home/mvkrein/etf_model/data'
etf_data_file = os.path.join(data_path,'etf_data_from_20120101.csv')
etf_data = pd.read_csv(etf_data_file,index_col=0)
```

```python
def lag(df, n):
    new_columns = ["{}_L{:02d}".format(variable, n) for variable in df.columns]
    new_df = df.shift(n)
    new_df.columns = new_columns
    return new_df
```

```python
lags = [5,10,21,42,63,84,126,189,252,-21]
for i,sym in enumerate(etf_list['Symbol']):
    etf_one_sym = etf_data.loc[etf_data['sym']==sym,]
    etf_one_sym.reset_index(inplace=True)
    for j,n in enumerate(lags):
        in_df = etf_one_sym.loc[:,['p','v']]
        new_df = lag(in_df,n)
        if j ==0:
            out_df = pd.concat([etf_one_sym,new_df],axis=1)
        else:
            out_df = pd.concat([out_df,new_df],axis=1)
    out_df['delta_p_L05'] = (out_df['p']/out_df['p_L05'])-1
    out_df['delta_p_L10'] = (out_df['p']/out_df['p_L10'])-1
    out_df['delta_p_L21'] = (out_df['p']/out_df['p_L21'])-1
    out_df['delta_p_L42'] = (out_df['p']/out_df['p_L42'])-1
    out_df['delta_p_L63'] = (out_df['p']/out_df['p_L63'])-1
    out_df['delta_p_L84'] = (out_df['p']/out_df['p_L84'])-1
    out_df['delta_p_L126'] = (out_df['p']/out_df['p_L126'])-1
    out_df['delta_p_L189'] = (out_df['p']/out_df['p_L189'])-1
    out_df['delta_p_L252'] = (out_df['p']/out_df['p_L252'])-1
    out_df['delta_p_L-21'] = (out_df['p_L-21']/out_df['p'])-1
    out_df['delta_v_L05'] = (out_df['v']/out_df['v_L05'])-1
    out_df['delta_v_L10'] = (out_df['v']/out_df['v_L10'])-1
    out_df['delta_v_L21'] = (out_df['v']/out_df['v_L21'])-1
    out_df['delta_v_L42'] = (out_df['v']/out_df['v_L42'])-1
    out_df['delta_v_L63'] = (out_df['v']/out_df['v_L63'])-1
    out_df['delta_v_L84'] = (out_df['v']/out_df['v_L84'])-1
    out_df['delta_v_L126'] = (out_df['v']/out_df['v_L126'])-1
    out_df['delta_v_L189'] = (out_df['v']/out_df['v_L189'])-1
    out_df['delta_v_L252'] = (out_df['v']/out_df['v_L252'])-1

    if i == 0:
        out_big_df = out_df
    else:
        out_big_df = pd.concat([out_big_df, out_df],axis=0)
out_big_df.sort_values(['Date','sym'],ascending=True,inplace=True)
```
